```python
from matplotlib import pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
Ages=[18, 20, 24, 30, 35, 50, 52, 56]
All_Developer=[52000, 55000, 58000, 60000, 63000, 68000, 72000, 75000]
plt.plot(Ages,All_Developer,color="k",marker='o',linestyle="--",label="All Developer",linewidth=0.5)
Python_Developer=[62000, 66000, 71000, 73000, 79000, 85000, 87000, 96000]
plt.plot(Ages,Python_Developer,color="b",marker=".",label="Python Developer",linewidth=0.5)
Javascript_Developer=[60000, 65000, 72000, 75000, 78000, 84000, 86000, 94000]
plt.plot(Ages,Javascript_Developer,color="r",label="Javascript Developer",linewidth=0.5)
plt.xlabel("Age")
plt.ylabel("Salary in USD")
plt.title("Salary in (USD) by Age")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
